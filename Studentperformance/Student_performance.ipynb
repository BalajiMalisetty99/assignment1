```python
import pandas as pd

# Load the dataset
df_students = pd.read_csv("StudentsPerformance.csv")

# Verify that the dataset is loaded
print(df_students.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for math, reading, and writing scores
plt.figure(figsize=(10, 5))
sns.histplot(df_students[['math score', 'reading score', 'writing score']], bins=20, kde=True)
plt.title("Distribution of Student Scores")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.legend(["Math", "Reading", "Writing"])
plt.show()
```

*Purpose*

This visualization shows how students' scores in Math, Reading, and Writing are distributed. It helps us understand the spread and frequency of scores.


*Analysis Made Easier*

Identifies the most common score ranges for each subject.
Reveals whether scores follow a normal distribution or have skewed distributions.
Highlights if some students consistently score lower or higher in all subjects.

```python
# Boxplot of test preparation vs math scores
plt.figure(figsize=(10, 5))
sns.boxplot(x="test preparation course", y="math score", hue="test preparation course", data=df_students, palette="coolwarm", dodge=False)
plt.title("Impact of Test Preparation Course on Math Scores")
plt.xlabel("Test Preparation Course")
plt.ylabel("Math Score")
plt.show()
```

Purpose

This boxplot compares Math Scores of students who completed vs did not complete the test preparation course.

Analysis Made Easier

Shows if students who took the preparation course had higher scores.
Reveals score variability between the two groups.
Helps in evaluating the effectiveness of test preparation programs.

```python
# Bar chart for average performance by gender
df_students.groupby('gender')[['math score', 'reading score', 'writing score']].mean().plot(kind='bar', figsize=(10, 5))
plt.title("Average Performance by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Score")
plt.legend(["Math", "Reading", "Writing"])
plt.show()
```

Purpose:

This bar chart compares average Math, Reading, and Writing scores between male and female students.

Analysis Made Easier:

Highlights which gender performs better in different subjects.
Helps in identifying if there are gender-based trends in academic performance.
Can inform educators on gender-based learning gaps and areas needing support.



```python
# Scatter plot for parental education vs math score
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df_students['parental level of education'], y=df_students['math score'], hue=df_students['gender'], palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Parental Education vs Math Performance")
plt.xlabel("Parental Level of Education")
plt.ylabel("Math Score")
plt.show()
```

Purpose:

This scatter plot examines whether a student’s math score is influenced by their parent's education level.

Analysis Made Easier:

Identifies if higher parental education correlates with higher math scores.
Helps determine whether family background impacts academic performance.
Useful for designing targeted academic support programs.

```python
# Heatmap for correlation between scores
plt.figure(figsize=(8, 6))
sns.heatmap(df_students[['math score', 'reading score', 'writing score']].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Scores")
plt.show()
```

Purpose:

This heatmap visualizes the correlation between Math, Reading, and Writing scores.

Analysis Made Easier:

Reveals strong positive correlations (e.g., students who do well in reading tend to do well in writing).
Helps identify cross-subject relationships that can guide teaching methods.
Useful for predicting scores in one subject based on performance in another.
